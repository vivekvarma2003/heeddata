```python
# Test Case 1: Concatenate productName and productId from 'azure_prices'
db = client['insider']
collection = db['azure_prices']
pipeline = [
    {
        '$project': {
            '_id': 0,
            'productDetails': {
                '$concat': ['$productName', ' - ', '$productId']
            }
        }
    }
]
documents = collection.aggregate(pipeline)

print("Transformation Test Case 1: Concatenated productName and productId")
for document in documents:
    print(document)
```

```python
# Test Case 2: Format migration_datetime from 'database_migrations'
db = client['insider']
collection = db['database_migrations']
pipeline = [
    {
        '$project': {
            '_id': 0,
            'migration_datetime': {
                '$dateFromString': {
                    'dateString': '$migration_datetime',
                    'format': "%Y%m%d%H%M%S"
                }
            }
        }
    }
]
documents = collection.aggregate(pipeline)

print("\nTransformation Test Case 2: Formatted migration_datetime")
for document in documents:
    print(document)
```

```python
# Test Case 3: Calculate time difference between migration_datetime and created_at in 'checklists'
db = client['restapi']
collection = db['checklists']
pipeline = [
    {
        '$project': {
            '_id': 0,
            'organization_id': 1,
            'created_at': 1,
            'migration_datetime': {
                '$dateFromString': {
                    'dateString': '$migration_datetime',
                    'format': "%Y%m%d%H%M%S"
                }
            },
            'time_difference_in_days': {
                '$divide': [
                    {
                        '$subtract': ['$created_at', {
                            '$dateFromString': {
                                'dateString': '$migration_datetime',
                                'format': "%Y%m%d%H%M%S"
                            }
                        }]
                    },
                    1000 * 60 * 60 * 24  # Convert milliseconds to days
                ]
            }
        }
    }
]
documents = collection.aggregate(pipeline)

print("\nTransformation Test Case 3: Time difference in days between migration_datetime and created_at")
for document in documents:
    print(document)
```
